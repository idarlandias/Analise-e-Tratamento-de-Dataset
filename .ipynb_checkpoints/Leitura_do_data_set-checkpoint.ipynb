```python
import pandas as pd

# Lê o arquivo diretamente da pasta local
df = pd.read_csv('fakeTelegram.BR_2022.csv')

# Limpa os nomes das colunas por segurança
df.columns = df.columns.str.strip()

print("Arquivo local lido com sucesso!")
print(df.head())
```

```python
# b) Identificar e listar as posições (células) com valores faltantes
# ------------------------------------------------------------------------------
print("\n--- Questão b) Posições com valores faltantes (primeiras 20) ---")
posicoes_faltantes = []
for col in df.columns:
    for i in df[df[col].isnull()].index:
        posicoes_faltantes.append((i, col))
print(posicoes_faltantes[:20]) # Mostra apenas as 20 primeiras para não poluir a saída
```

```python
# c) Contar quantas linhas possuem valores faltantes
# ------------------------------------------------------------------------------
print("\n--- Questão c) Número de linhas com valores faltantes ---")
linhas_com_faltantes = df.isnull().any(axis=1).sum()
print(f"O número de linhas com valores faltantes é: {linhas_com_faltantes}")
```

```python
# d) Contar valores faltantes por coluna (feature)
# ------------------------------------------------------------------------------
print("\n--- Questão d) Quantidade de valores faltantes por coluna ---")
faltantes_por_coluna = df.isnull().sum()
print(faltantes_por_coluna)
```

```python
# e) Identificar e listar as linhas repetidas (duplicadas)
 # ------------------------------------------------------------------------------
print("\n--- Questão e) Linhas duplicadas ---")
linhas_duplicadas = df[df.duplicated()]
print(linhas_duplicadas)
```

```python
# f) Identificar valores que não pertencem ao domínio esperado
# ------------------------------------------------------------------------------
print("\n--- Questão f) Posições com valores fora do domínio esperado ---")
posicoes_fora_dominio = []
colunas_numericas = ['message_id', 'sender_id', 'reply_to_message_id', 'views', 'forwards']

for col in colunas_numericas:
    if col in df.columns:
        for i, val in df[col].items():
            if not pd.api.types.is_number(val) and pd.notna(val):
                posicoes_fora_dominio.append((i, col, val))

if 'date' in df.columns:
    for i, val in df['date'].items():
        try:
            pd.to_datetime(val)
        except (ValueError, TypeError):
            if pd.notna(val):
                posicoes_fora_dominio.append((i, 'date', val))

print(posicoes_fora_dominio)
```

```python
# CÉLULA DE VERIFICAÇÃO
print("--- Verificando as colunas do DataFrame 'df' ---")
print("As colunas atualmente na memória são:")
print(df.columns.tolist())
```

```python
# g) Criar a coluna "caracteres"
# ------------------------------------------------------------------------------
print("\n--- Questão g) Criando a coluna 'caracteres' ---")
# CORREÇÃO: Usando 'text_content_anonymous'
df['caracteres'] = df['text_content_anonymous'].str.len().fillna(0).astype(int)
print(df[['text_content_anonymous', 'caracteres']].head())
```

```python
# h) Criar a coluna "words"
# ------------------------------------------------------------------------------
print("\n--- Questão h) Criando a coluna 'words' ---")
# CORREÇÃO: Usando 'text_content_anonymous'
df['words'] = df['text_content_anonymous'].str.split().str.len().fillna(0).astype(int)
print(df[['text_content_anonymous', 'words']].head())
```

```python
# i) Criar a coluna "viral"
# ------------------------------------------------------------------------------
print("\n--- Questão i) Criando a coluna 'viral' ---")
# CORREÇÃO: Usando 'text_content_anonymous'
df['viral'] = df['text_content_anonymous'].duplicated(keep=False).astype(int)
print(df[['text_content_anonymous', 'viral']].head())
```

```python
# j) Criar a coluna "sharings"
# ------------------------------------------------------------------------------
print("\n--- Questão j) Criando a coluna 'sharings' ---")
# CORREÇÃO: Usando 'text_content_anonymous'
df['sharings'] = df['text_content_anonymous'].map(df['text_content_anonymous'].value_counts()).fillna(0).astype(int)
print(df[['text_content_anonymous', 'sharings']].head())
```

```python
# k) Criar a coluna "sentiment"
# ------------------------------------------------------------------------------
print("\n--- Questão k) Criando a coluna 'sentiment' ---")
def classify_sentiment(text):
    if isinstance(text, str):
        text_lower = text.lower()
        positive_keywords = ['bom', 'ótimo', 'excelente', 'feliz', 'sucesso', 'amor', 'bom dia']
        negative_keywords = ['ruim', 'péssimo', 'terrível', 'triste', 'problema', 'ódio', 'cuidado', 'golpe']
        positive_count = sum(word in text_lower for word in positive_keywords)
        negative_count = sum(word in text_lower for word in negative_keywords)
        if positive_count > negative_count: return 1
        elif negative_count > positive_count: return -1
        else: return 0
    return 0
# CORREÇÃO: Usando 'text_content_anonymous'
df['sentiment'] = df['text_content_anonymous'].apply(classify_sentiment)
print(df[['text_content_anonymous', 'sentiment']].head())
```

```python
# l) Eliminar linhas com "trava-zaps"
# ------------------------------------------------------------------------------
print("\n--- Questão l) Eliminando linhas com 'trava-zaps' ---")
linhas_antes = len(df)
# CORREÇÃO: Usando a coluna correta 'trava_zap' que já existe
df = df[df['trava_zap'] == False] 
linhas_depois = len(df)
print(f"Linhas antes: {linhas_antes}, Linhas depois: {linhas_depois}. Total removido: {linhas_antes - linhas_depois}")
```

```python
# m) Identificar inconsistências entre os atributos
# ------------------------------------------------------------------------------
print("\n--- Questão m) Identificando inconsistências ---")

# Para as inconsistências, precisamos verificar se as colunas existem
colunas_necessarias = ['id_message', 'score_sentiment', 'date_message']
if all(col in df.columns for col in colunas_necessarias):
    # Inconsistência 1: Exemplo - Score de sentimento vs. Análise de sentimento
    # Esta é uma análise mais avançada, apenas um exemplo
    print("\nInconsistência 1: Comparando scores de sentimento (exemplo):")
    print(df[['score_sentiment', 'sentiment']].head())

    # Inconsistência 2: Datas no futuro
    df['date_message'] = pd.to_datetime(df['date_message'], errors='coerce') 
    from datetime import datetime
    agora = datetime.now()
    datas_futuras = df[df['date_message'] > agora]
    print("\nInconsistência 2: Mensagens com data no futuro:")
    print(datas_futuras[['id_message', 'date_message']])
else:
    print("\nAlgumas colunas necessárias para a análise de inconsistências não foram encontradas.")
    
print("\n--- ANÁLISE COMPLETA CONCLUÍDA ---")
```
